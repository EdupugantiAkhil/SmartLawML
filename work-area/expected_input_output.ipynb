```python
#cases 1 from given statement find given section and act 
for i in dataset['train']['background']:
    print("input:" + i['text'])
    print("expected output:"  )
    for j in i['legalReferences']:
        print(j['legalActFound']['stringMatched'])
    break
    
```

```python
#case 2 Given an argument-text, detect if this argument is by applicant-lawyer, respondent-lawyer or judge.
for i in dataset['train']['arguments']:
    for j in i:
        print(j['argumentNumber'])
        print("input: " + j['text'])
        print('output: ' + j['argumentBy'])
        break 
    break
    
```

```python
#case 3 Given a sentence-text, predict if it is premise or conclusion

for i in dataset['train']['arguments']:
    for j in i:
        for k in j['argumentSentences']:
            print('input:'+ str(k['sentenceNumber']))
            print(k['text'])
            print('output: ' + k['argumentSentenceType'])
        break
        
    break
```
